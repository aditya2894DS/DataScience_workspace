<p>Let's check if their is null or unique values in our dataframe</p>

```python
def null_columns(df):
  null_cols = []

  for col in df:
    if df[col].isnull().sum() != 0:
      null_cols.append(col)

  print("There are {} columns having null values: {}".format(len(null_cols), null_cols))

null_columns(joined_df)
```

<p>Let's again resplit the into training and testing datasets.</p>

```python
train_df1 = joined_df.merge(train_df, on='ID', how='inner')
test_df1 = joined_df.merge(test_df, on='ID', how='inner')
```

<p>Let's start our data preparation procedure on our training dataset</p>

<p>Finding columns with suffix '_y'</p>

```python
y_cols = [col for col in train_df1.columns if '_y' in col]

train_df1  = train_df1.drop(y_cols, axis=1)
test_df1 = test_df1.drop(y_cols, axis=1)
```

<p>Let's separate the object columns from both the datasets</p>

```python
train_obj_df = train_df1.select_dtypes('object')
test_obj_df = test_df1.select_dtypes('object')
```

<p>Now we have our training data</p>

<p>Let's use label encoding for our object columns in our training dataset. In case we find an new variable in test dataset while transforming, then we will encode it as 'unknown'.</p>

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

lab_encoder = LabelEncoder()
x0 = lab_encoder.fit(train_obj_df['X0_x'])
test_obj_df['X0_x'] = test_obj_df['X0_x'].map(lambda s: '<unknown>' if s not in x0.classes_ else s)
x0.classes_ = np.append(x0.classes_, '<unknown>')
encoded_train_x0 = x0.transform(train_obj_df['X0_x'])
encoded_test_x0 = x0.transform(test_obj_df['X0_x'])

x1 = lab_encoder.fit(train_obj_df['X1_x'])
test_obj_df['X1_x'] = test_obj_df['X1_x'].map(lambda s: 'unknown' if s not in x1.classes_ else s)
x1.classes_ = np.append(x1.classes_, '<unknown>')
encoded_train_x1 = x1.transform(train_obj_df['X1_x'])
encoded_test_x1 = x1.transform(test_obj_df['X1_x'])

x2 = lab_encoder.fit(train_obj_df['X2_x'])
test_obj_df['X2_x'] = test_obj_df['X2_x'].map(lambda s: '<unknown>' if s not in x2.classes_ else s)
x2.classes_ = np.append(x2.classes_, '<unknown>')
encoded_train_x2 = x2.transform(train_obj_df['X2_x'])
encoded_test_x2 = x2.transform(test_obj_df['X2_x'])

x3 = lab_encoder.fit(train_obj_df['X3_x'])
test_obj_df['X3_x'] = test_obj_df['X3_x'].map(lambda s: '<unknown>' if s not in x3.classes_ else s)
x3.classes_ = np.append(x3.classes_, '<unknown>')
encoded_train_x3 = x3.transform(train_obj_df['X3_x'])
encoded_test_x3 = x3.transform(test_obj_df['X3_x'])

x4 = lab_encoder.fit(train_obj_df['X4_x'])
test_obj_df['X4_x'] = test_obj_df['X4_x'].map(lambda s: '<unknown>' if s not in x4.classes_ else s)
x4.classes_ = np.append(x4.classes_, '<unknown>')
encoded_train_x4 = x4.transform(train_obj_df['X4_x'])
encoded_test_x4 = x4.transform(test_obj_df['X4_x'])

x5 = lab_encoder.fit(train_obj_df['X5_x'])
test_obj_df['X5_x'] = test_obj_df['X5_x'].map(lambda s: '<unknown>' if s not in x5.classes_ else s)
x5.classes_ = np.append(x5.classes_, '<unknown>')
encoded_train_x5 = x5.transform(train_obj_df['X5_x'])
encoded_test_x5 = x5.transform(test_obj_df['X5_x'])

x6 = lab_encoder.fit(train_obj_df['X6_x'])
test_obj_df['X6_x'] = test_obj_df['X6_x'].map(lambda s: '<unknown>' if s not in x6.classes_ else s)
x6.classes_ = np.append(x6.classes_, '<unknown>')
encoded_train_x6 = x6.transform(train_obj_df['X6_x'])
encoded_test_x6 = x6.transform(test_obj_df['X6_x'])

x8 = lab_encoder.fit(train_obj_df['X8_x'])
test_obj_df['X8_x'] = test_obj_df['X8_x'].map(lambda s: '<unknown>' if s not in x8.classes_ else s)
x8.classes_ = np.append(x8.classes_, '<unknown>')
encoded_train_x8 = x8.transform(train_obj_df['X8_x'])
encoded_test_x8 = x8.transform(test_obj_df['X8_x'])
```

```python
encoded_train_values = np.array([encoded_train_x0, encoded_train_x1, 
              encoded_train_x2, encoded_train_x3, 
              encoded_train_x4, encoded_train_x5,
              encoded_train_x6, encoded_train_x8])

encoded_test_values = np.array([encoded_test_x0, encoded_test_x1, 
              encoded_test_x2, encoded_test_x3, 
              encoded_test_x4, encoded_test_x5,
              encoded_test_x6, encoded_test_x8])

encoded_train_values = encoded_train_values.T
encoded_test_values = encoded_test_values.T
```

```python
encoded_train_df = pd.DataFrame(data=encoded_train_values, columns=train_obj_df.columns)
encoded_test_df = pd.DataFrame(data=encoded_test_values, columns=test_obj_df.columns)
```

```python
train_num_df = train_df1.select_dtypes(exclude='object')
test_num_df = test_df1.select_dtypes(exclude='object')
train_num_df
```

<p>Merging encoded data with train data</p>

```python
final_train_df = pd.merge(encoded_train_df, train_num_df, left_index=True, right_index=True)
final_test_df = pd.merge(encoded_test_df, test_num_df, left_index=True, right_index=True)
```

<p>Merging encoded data with train data</p>

```python
from sklearn.preprocessing import StandardScaler

pca_df = final_train_df.drop(['ID', 'y'], axis=1)
pca_test_df = final_test_df.drop('ID', axis=1)

scaler = StandardScaler()
pca_df_scaler = scaler.fit(pca_df)
pca_df_values = pca_df_scaler.transform(pca_df)

pca_df = pd.DataFrame(columns=pca_df.columns, data=pca_df_values)
pca_df2 = pca_df.copy()

pca_df.head()
```

<p>Let's start factor analysis of our training set. For this we will use PCA.</p>

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(pca_df)

pca_df_trans = pca.transform(pca_df)

pca_df = pd.DataFrame(columns=pca_df.columns, data=pca_df_trans)

pca_df.head()
```

```python
component_names = [f'PCA_{i+1}' for i in range(pca_df.shape[1])]
pca_loading_df = pd.DataFrame(index=component_names, columns=pca_df.columns, data=pca.components_)

pca_loading_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  
sns.set_theme(style='whitegrid')

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
axs = axs.flatten()

sns.barplot(y=pca_df.columns[0:20][::-1], x=pca.explained_variance_[0:20][::-1], ax=axs[0])
sns.barplot(y=pca_df.columns[0:20][::-1], x=pca.explained_variance_ratio_[0:20][::-1], ax=axs[1])
plt.show()
```

<p>As you can see we have the explained variance by all the principal components. We will consider top 7 components as our features.</p>

```python
pca_final = PCA(n_components=7).fit(pca_df2)
pca_df2_values = pca_final.transform(pca_df2)

pca_final_test = pca_final.transform(pca_test_df)
```

<p>Let's use XG boost for modelling. We will also do hyperparameter tuning using randomized search CV</p>

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb

xg_reg = xgb.XGBRegressor()
```

<p>Let's create a dictionary with keys as the parameters and values will be arrays of values</p>

```python
params = {
 'learning_rate' : [0.05,0.10,0.15,0.20,0.25,0.30],
 'max_depth' : [ 3, 4, 5, 6, 8, 10, 12, 15],
 'min_child_weight' : [ 1, 3, 5, 7 ],
 'gamma': [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 'colsample_bytree' : [ 0.3, 0.4, 0.5 , 0.7 ]
}
```

```python
model = RandomizedSearchCV(xg_reg, param_distributions=params, n_iter=5, cv=10, n_jobs=-1, verbose=1)
model.fit(pca_df2_values, final_train_df['y'])
```

<p>Our model is fitted using combinations of various parameters. Now let's look which combination of params. were the most optimized.</p>

```python

model.best_params_ 
```

<p>Now we will rerun the XGBoost model with the above mentioned parameters</p>

```python
final_model = xgb.XGBRegressor(min_child_weight=1,
                               max_depth=10,
                               learning_rate=0.05,
                               gamma=0.1,
                               colsample_bytree=0.7)
final_model.fit(pca_df2_values, final_train_df['y'])
```

```python
pred = final_model.predict(pca_final_test)
pred
```
